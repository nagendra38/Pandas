```python
import pandas as pd
temp_df=pd.DataFrame({
    'city' : ['mumbai', 'delhi', 'banglore','raipur'],
    'temperature' : [32,45,30,34]
},index=[0,1,2,4])
temp_df
```

```python
wind_df=pd.DataFrame({
    'city' : ['mumbai', 'delhi', 'banglore','rajnandgaon'],
    'humidity' : [80,60,78,65]
},index=[0,1,2,3])
wind_df
```

```python
df3=pd.merge(temp_df,wind_df,on='city')
df3
```

```python
df3=pd.merge(temp_df,wind_df,on='city')
df3
```

```python
df3=pd.merge(temp_df,wind_df,on='city',how='outer')
df3
```

```python
df3=pd.merge(temp_df,wind_df,on='city',how='left')
df3
```

```python
df3=pd.merge(temp_df,wind_df,on='city',how='right')
df3
```

```python
df3=pd.merge(temp_df,wind_df,on='city',how='outer',indicator=True)
df3
```
